**CIRCUITO CON UN CÚBIT**

```python
!pip install qiskit

import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

# Valores de theta de 0 a 2*pi
thetas = np.linspace(0, 2 * np.pi, 100)

# Listas para guardar las probabilidades
prob_0 = []
prob_1 = []

for theta in thetas:
    # Crear circuito de 1 cúbit
    qc = QuantumCircuit(1)

    # Aplicar la puerta Ry
    qc.ry(theta, 0)

    # Obtener el vector de estado exacto
    state = Statevector.from_instruction(qc)
    probs = state.probabilities()

    # Guardar las probabilidades P(0) y P(1)
    prob_0.append(probs[0])
    prob_1.append(probs[1])

# Graficar los resultados
plt.figure(figsize=(8, 5))
plt.plot(thetas, prob_0, label='P(0)', color='blue', linewidth=2)
plt.plot(thetas, prob_1, label='P(1)', color='red', linewidth=2)
plt.title('Evolución de Probabilidades para 1 Cúbit con puerta Ry(θ)')
plt.xlabel('Theta (radianes)')
plt.ylabel('Probabilidad')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['0', 'π/2', 'π', '3π/2', '2π'])
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

**CIRCUITO CON DOS CÚBITS**

```python
# Listas para guardar las 4 probabilidades
prob_00 = []
prob_01 = []
prob_10 = []
prob_11 = []

for theta in thetas:
    # Crear circuito de 2 cúbits
    qc2 = QuantumCircuit(2)

    # Aplicar Ry a ambos cúbits
    qc2.ry(theta, 0)
    qc2.ry(theta, 1)

    # Obtener el vector de estado
    state2 = Statevector.from_instruction(qc2)
    probs2 = state2.probabilities()

    # Guardar probabilidades
    prob_00.append(probs2[0]) # |00>
    prob_01.append(probs2[1]) # |01>
    prob_10.append(probs2[2]) # |10>
    prob_11.append(probs2[3]) # |11>

# Graficar los resultados de 2 cúbits
plt.figure(figsize=(8, 5))
plt.plot(thetas, prob_00, label='P(00)', linewidth=2)
plt.plot(thetas, prob_01, label='P(01)', linewidth=2, linestyle='--')
plt.plot(thetas, prob_10, label='P(10)', linewidth=2, linestyle=':')
plt.plot(thetas, prob_11, label='P(11)', linewidth=2)
plt.title('Evolución de Probabilidades para 2 Cúbits con puertas Ry(θ)')
plt.xlabel('Theta (radianes)')
plt.ylabel('Probabilidad')
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['0', 'π/2', 'π', '3π/2', '2π'])
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

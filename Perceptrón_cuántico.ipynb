# EJERCICIO 1

Representación de la función f(x)=x^2 utilizando un único perceptrón cuántico sinuidal, y con función de pérdida error cuadrático medio (MSE).

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Preparación de los datos
x = np.linspace(-1, 1, 100)
y_true = x**2

# 2. Definición de los Parámetros
w = 1.0
b = 0.0

# Hiperparámetros
lr = 0.05
epochs = 800

# 3. Ciclo de Entrenamiento
for epoch in range(epochs):
    # A. FORWARD PASS: calculamos las predicciones
    y_pred = (np.sin((w * x + b) / 2))**2

    # B. CÁLCULO DE LA PÉRDIDA: Error Cuadrático Medio (MSE)
    loss = np.mean((y_pred - y_true)**2)

    # C. BACKWARD PASS: Cálculo manual de los gradientes
    error = y_pred - y_true

    # Aplicamos las fórmulas de las derivadas calculadas arriba
    grad_w = np.mean(error * np.sin(w * x + b) * x)
    grad_b = np.mean(error * np.sin(w * x + b))

    # D. ACTUALIZACIÓN DE LOS PESOS
    # Restamos el gradiente multiplicado por la tasa de aprendizaje (learning rate)
    w -= lr * grad_w
    b -= lr * grad_b

    # Imprimimos el progreso cada 100 épocas
    if (epoch + 1) % 100 == 0:
        print(f"Época {epoch+1}/{epochs} | Loss: {loss:.6f} | w: {w:.4f} | b: {b:.4f}")

print("\n¡Entrenamiento completado!")

# 4. Visualización de los Resultados
y_pred_final = (np.sin((w * x + b) / 2))**2
plt.figure(figsize=(8, 6))
plt.plot(x, y_true, label='x^2', color='tab:blue')
plt.plot(x, y_pred_final, label='modelo', color='tab:orange')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left')
plt.title('Ejercicio 1: f(x) = x^2 con Perdida MSE')
plt.show()
```

# EJERCICIO 2

Representación de la función f(x)=x^2 utilizando un único perceptrón cuántico sinuidal, y con función de pérdida entropía cruzada.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definir los datos de entrenamiento
x = np.linspace(-1, 1, 100)
y_true = x**2

# 2. Inicialización de parámetros
w = 1.0
b = 0.0

# 3. Hiperparámetros
epochs = 800
learning_rate = 0.01
eps = 1e-8            # Para evitar divisiones por cero

# 4. Bucle de entrenamiento con ENTROPÍA CRUZADA
for epoch in range(epochs):
    # Forward pass: Función de activación cuántica
    y_pred = np.sin((w * x + b) / 2)**2

    # Recortamos por seguridad matemática
    y_pred_clipped = np.clip(y_pred, eps, 1 - eps)

    # Backward pass: Derivadas de la Entropía Cruzada
    grad_loss_y_pred = -( (y_true / y_pred_clipped) - ((1 - y_true) / (1 - y_pred_clipped)) )

    dy_pred_dw = (x / 2) * np.sin(w * x + b)
    dy_pred_db = (1 / 2) * np.sin(w * x + b)

    grad_w = np.mean(grad_loss_y_pred * dy_pred_dw)
    grad_b = np.mean(grad_loss_y_pred * dy_pred_db)

    # Actualizar los pesos
    w -= learning_rate * grad_w
    b -= learning_rate * grad_b

# 5. Imprimir resultados y visualizar
print("Entrenamiento Ejercicio 2 finalizado.")
print(f"w = {w:.15f}")
print(f"b = {b:.15f}")

plt.figure(figsize=(8, 6))
plt.plot(x, y_true, label='x^2', color='tab:blue')
plt.plot(x, np.sin((w * x + b) / 2)**2, label='modelo', color='tab:orange')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left')
plt.title('Ejercicio 2: f(x) = x^2 con Entropía Cruzada')
plt.show()
```

Función de pérdida entropía cruzada

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Preparación de los datos
x = np.linspace(-1, 1, 100)
# Definición de la función objetivo: Sigmoide
y_true = 1 / (1 + np.exp(-x))

# 2. Definición de los Parámetros
w = 1.0
b = 0.0

# Hiperparámetros
lr = 0.05
epochs = 1000

# 3. Ciclo de Entrenamiento
for epoch in range(epochs):
    # A. FORWARD PASS: Calculamos las predicciones
    y_pred = (np.sin((w * x + b) / 2))**2

    # Aplicación de límite de seguridad (épsilon) para evitar log(0) o división por cero
    eps = 1e-7
    y_pred_safe = np.clip(y_pred, eps, 1 - eps)

    # B. CÁLCULO DE LA PÉRDIDA: Entropía Cruzada Binaria (BCE)
    # Fórmula: -media(y * log(y_pred) + (1 - y) * log(1 - y_pred))
    loss = -np.mean(y_true * np.log(y_pred_safe) + (1 - y_true) * np.log(1 - y_pred_safe))

    # C. BACKWARD PASS: Cálculo de gradientes usando la regla de la cadena

    # Paso 1: Derivada de la pérdida BCE respecto a la predicción
    dL_dypred = (y_pred_safe - y_true) / (y_pred_safe * (1 - y_pred_safe))

    # Paso 2: Derivada de la predicción respecto a los parámetros w y b
    # Nota: La derivada analítica de sin^2((wx+b)/2) incluye la derivada interna
    dypred_dw = 0.5 * np.sin(w * x + b) * x
    dypred_db = 0.5 * np.sin(w * x + b)

    # Paso 3: Multiplicación para obtener los gradientes finales
    grad_w = np.mean(dL_dypred * dypred_dw)
    grad_b = np.mean(dL_dypred * dypred_db)

    # D. ACTUALIZACIÓN DE LOS PESOS
    # Restamos el gradiente multiplicado por la tasa de aprendizaje (learning rate)
    w -= lr * grad_w
    b -= lr * grad_b

    # Imprimimos el progreso cada 100 épocas
    if (epoch + 1) % 100 == 0:
        print(f"Época {epoch+1}/{epochs} | Loss BCE: {loss:.6f} | w: {w:.4f} | b: {b:.4f}")

print("\n¡Entrenamiento completado!")

# 4. Visualización de los Resultados
y_pred_final = (np.sin((w * x + b) / 2))**2
plt.figure(figsize=(8, 6))

# Graficamos la función real y la predicción del modelo
plt.plot(x, y_true, label='Sigmoide Real', color='tab:blue')
plt.plot(x, y_pred_final, label='Modelo Entrenado (BCE)', color='tab:orange', linestyle='--')

# Configuración de los ejes y formato visual
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.title('Aproximación de la Sigmoide con Pérdida de Entropía Cruzada')
plt.grid(True, alpha=0.3)
plt.show()
```

# EJERCICIO 4

Representación de la función f(x)=x utilizando un perceptrón cuántico que combina seno y coseno.

Función de pérdida error cuadrático medio.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Preparación de los datos
x = np.linspace(-2, 2, 100)
# Definición de la función objetivo: f(x) = x
y_true = x

# 2. Definición de los Parámetros
w = 1.0
b = 0.0
# Inicialización de los parámetros C y D en los extremos esperados
c = 2.0
d = -2.0

# Hiperparámetros
lr = 0.03
epochs = 1500

# 3. Ciclo de Entrenamiento
for epoch in range(epochs):
    # A. FORWARD PASS: Cálculo de las probabilidades de inspiración cuántica
    theta = w * x + b
    p1 = (np.sin(theta / 2))**2
    p0 = (np.cos(theta / 2))**2

    # Construcción de la predicción combinando probabilidades y pesos
    y_pred = c * p1 + d * p0

    # B. CÁLCULO DE LA PÉRDIDA: Error Cuadrático Medio (MSE)
    loss = np.mean((y_pred - y_true)**2)

    # C. BACKWARD PASS: Cálculo manual de los gradientes
    error = y_pred - y_true

    # Derivadas de la pérdida respecto a los parámetros C y D
    grad_c = np.mean(2 * error * p1)
    grad_d = np.mean(2 * error * p0)

    # Derivadas de la pérdida respecto a los parámetros w y b usando la regla de la cadena
    grad_w = np.mean(2 * error * 0.5 * (c - d) * np.sin(theta) * x)
    grad_b = np.mean(2 * error * 0.5 * (c - d) * np.sin(theta))

    # D. ACTUALIZACIÓN DE LOS PESOS
    # Restamos el gradiente multiplicado por la tasa de aprendizaje (learning rate)
    c -= lr * grad_c
    d -= lr * grad_d
    w -= lr * grad_w
    b -= lr * grad_b

    # Imprimimos el progreso cada 200 épocas
    if (epoch + 1) % 200 == 0:
        print(f"Época {epoch+1}/{epochs} | Loss: {loss:.4f} | w: {w:.2f} | b: {b:.2f} | C: {c:.2f} | D: {d:.2f}")

print("\n¡Entrenamiento completado!")

# 4. Visualización de los Resultados
theta_final = w * x + b
y_pred_final = c * (np.sin(theta_final / 2))**2 + d * (np.cos(theta_final / 2))**2

plt.figure(figsize=(8, 6))

# Graficamos la función real y la predicción del modelo
plt.plot(x, y_true, label='f(x) = x', color='tab:blue')
plt.plot(x, y_pred_final, label='Modelo A(x) (Simulación Cuántica)', color='tab:orange', linestyle='--')

# Configuración de los ejes y formato visual
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.title('Aproximación de f(x)=x con estructura A(x) = C P(|1>) + D P(|0>)')
plt.grid(True, alpha=0.3)
plt.show()
```

Función de pérdida entropía cruzada

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Preparación de los datos
x = np.linspace(-2, 2, 100)
# Definición de la función objetivo original: f(x) = x
y_true_original = x

# Normalización de los datos: Conversión del rango [-2, 2] al rango [0, 1]
# Fórmula: (valor - min) / (max - min)
y_true_norm = (x + 2) / 4

# 2. Definición de los Parámetros
w = 1.0
b = 0.0
# Inicialización de los parámetros C y D en el rango [0, 1] para la salida probabilística
c = 0.8
d = 0.2

# Hiperparámetros
lr = 0.05
epochs = 1500

# 3. Ciclo de Entrenamiento
for epoch in range(epochs):
    # A. FORWARD PASS: Cálculo de las probabilidades de inspiración cuántica
    theta = w * x + b
    p1 = (np.sin(theta / 2))**2
    p0 = (np.cos(theta / 2))**2

    # Construcción de la predicción combinando probabilidades y pesos
    y_pred = c * p1 + d * p0

    # Aplicación de límite de seguridad (épsilon) para evitar log(0) o división por cero en BCE
    eps = 1e-7
    y_pred_safe = np.clip(y_pred, eps, 1 - eps)

    # B. CÁLCULO DE LA PÉRDIDA: Entropía Cruzada Binaria (BCE) con datos normalizados
    loss = -np.mean(y_true_norm * np.log(y_pred_safe) + (1 - y_true_norm) * np.log(1 - y_pred_safe))

    # C. BACKWARD PASS: Cálculo de gradientes usando la regla de la cadena

    # Paso 1: Derivada de la pérdida BCE respecto a la predicción
    dL_dypred = (y_pred_safe - y_true_norm) / (y_pred_safe * (1 - y_pred_safe))

    # Paso 2: Derivada de la predicción respecto a los parámetros C y D
    grad_c = np.mean(dL_dypred * p1)
    grad_d = np.mean(dL_dypred * p0)

    # Paso 3: Derivada de la predicción respecto a los parámetros w y b
    grad_w = np.mean(dL_dypred * 0.5 * (c - d) * np.sin(theta) * x)
    grad_b = np.mean(dL_dypred * 0.5 * (c - d) * np.sin(theta))

    # D. ACTUALIZACIÓN DE LOS PESOS
    # Restamos el gradiente multiplicado por la tasa de aprendizaje (learning rate)
    c -= lr * grad_c
    d -= lr * grad_d
    w -= lr * grad_w
    b -= lr * grad_b

    # Imprimimos el progreso cada 200 épocas
    if (epoch + 1) % 200 == 0:
        print(f"Época {epoch+1}/{epochs} | Loss BCE: {loss:.4f} | w: {w:.2f} | b: {b:.2f} | C: {c:.2f} | D: {d:.2f}")

print("\n¡Entrenamiento completado!")

# 4. Visualización de los Resultados
theta_final = w * x + b
# Predicción en la escala normalizada [0, 1]
y_pred_norm = c * (np.sin(theta_final / 2))**2 + d * (np.cos(theta_final / 2))**2

# Desnormalización de los datos: Reversión de la predicción a la escala original [-2, 2]
# Fórmula inversa: valor_norm * (max - min) + min
y_pred_final = y_pred_norm * 4 - 2

plt.figure(figsize=(8, 6))

# Graficamos la función real original y la predicción desnormalizada del modelo
plt.plot(x, y_true_original, label='f(x) = x', color='tab:blue')
plt.plot(x, y_pred_final, label='Modelo Cuántico (Entrenado con BCE)', color='tab:orange', linestyle='--')

# Configuración de los ejes y formato visual
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.title('Aproximación de f(x)=x con BCE (Normalizando datos)')
plt.grid(True, alpha=0.3)
plt.show()
```

# EJERCICIO 5

Representación de la función f(x)=x(1-x^2) utilizando un perceptrón cuántico que combina seno y coseno.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definir los datos de entrenamiento
x = np.linspace(-1, 1, 100)
y_true = x * (1 - x**2)

# 2. Inicialización de parámetros
w = 2.5
b = 1.5
C = 1.0
D = 0.0

# 3. Hiperparámetros
epochs = 500
learning_rate = 0.05

# 4. Bucle de entrenamiento con ERROR CUADRÁTICO MEDIO (MSE)
for epoch in range(epochs):
    # Forward pass
    y_pred = (C + D) / 2 - ((C - D) / 2) * np.cos(w * x + b)

    # Backward pass
    grad_loss_y_pred = y_pred - y_true

    # Derivadas respecto a los 4 parámetros
    dy_pred_dw = ((C - D) / 2) * x * np.sin(w * x + b)
    dy_pred_db = ((C - D) / 2) * np.sin(w * x + b)
    dy_pred_dC = 0.5 - 0.5 * np.cos(w * x + b)
    dy_pred_dD = 0.5 + 0.5 * np.cos(w * x + b)

    # Regla de la cadena
    grad_w = np.mean(grad_loss_y_pred * dy_pred_dw)
    grad_b = np.mean(grad_loss_y_pred * dy_pred_db)
    grad_C = np.mean(grad_loss_y_pred * dy_pred_dC)
    grad_D = np.mean(grad_loss_y_pred * dy_pred_dD)

    # Actualizar los pesos
    w -= learning_rate * grad_w
    b -= learning_rate * grad_b
    C -= learning_rate * grad_C
    D -= learning_rate * grad_D

# 5. Visualizar
plt.figure(figsize=(7, 5))
plt.plot(x, y_true, label='x(1 - x^2)', color='tab:blue')

# Recalculamos la linea con los pesos finales
y_final = (C + D) / 2 - ((C - D) / 2) * np.cos(w * x + b)
plt.plot(x, y_final, label='modelo', color='tab:orange')

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper center')
plt.title('N = 500')
plt.show()
```
